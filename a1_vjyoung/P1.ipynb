## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    vert_offset = 0.6
    height, width, _ = img.shape
    
    left_x = []
    left_y = []
    left_weights = []
    right_x = []
    right_y = []
    right_weights = []

    # Classify lines into left and right (using slope and which half of the image the lines are on)
    for line in lines:
        for x1, y1, x2, y2 in line:
            slope = (y2-y1)/(x2-x1)
            length = np.sqrt((y2-y1)**2 + (x2-x1)**2)
            if slope >= 0 and (x1>(width/2) or x2>(width/2)):
                #right.append(line)
                right_x.extend([x1, x2])
                right_y.extend([y1, y2])
                right_weights.extend([length, length])
            elif slope < 0 and (x1<(width/2) or x2<(width/2)):
                #left.append(line)
                left_x.extend([x1, x2])
                left_y.extend([y1, y2])
                left_weights.extend([length, length])

    # Polyfit to group left and right lines using lengths of lines as weights.
    y1 = height
    y2 = int(vert_offset*height)

    lanes = []
    
    if left_x != []:
        m, b = np.polyfit(left_x, left_y, 1, w=left_weights)
        x1_left = int((y1-b)/m)
        x2_left = int((y2-b)/m)
        left_lane = [[x1_left, y1, x2_left, y2]]
        lanes.append(left_lane)

    if right_x != []:
        m, b = np.polyfit(right_x, right_y, 1, w=right_weights)
        x1_right = int((y1-b)/m)
        x2_right = int((y2-b)/m)
        right_lane = [[x1_right, y1, x2_right, y2]]
        lanes.append(right_lane)

    for lane in lanes:
        for x1,y1,x2,y2 in lane:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
in_dir = "test_images/"
out_dir = "test_images_output/"
report_dir = "report_images/"

# Iterate through images in directory
for image_name in os.listdir(in_dir):
    #image_name = "solidWhiteCurve.jpg"

    image = mpimg.imread(in_dir+image_name)


    # Convert image to greyscale
    image_conv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(image_conv)

    # Threshold greyscale image
    v = (v > 200) * v

    height, width = v.shape

    plt.figure()
    plt.imshow(v, cmap='gray')
    mpimg.imsave(report_dir + image_name.split('.')[0] + '_1vchannel.jpg', v, cmap = cm.gray)

    # Gaussian blur
    v = gaussian_blur(v, 7)
    plt.figure()
    plt.imshow(v, cmap='gray')
    mpimg.imsave(report_dir + image_name.split('.')[0] + '_2gauss.jpg', v, cmap = cm.gray)

    # Perform dilation to make sure edges are continuous
    v = cv2.dilate(v, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7)), iterations = 1)

    plt.figure()
    plt.imshow(v, cmap='gray')
    mpimg.imsave(report_dir + image_name.split('.')[0] + '_3dilation.jpg', v, cmap = cm.gray)

    # Canny edge detector
    edges = canny(v, 100, 200)

    plt.figure()
    plt.imshow(edges, cmap='gray')
    mpimg.imsave(report_dir + image_name.split('.')[0] + '_4canny.jpg', edges, cmap = cm.gray)

    # ROI (trapezoid)
    horiz_offset_bottom = 0.05
    horiz_offset_top = 0.35
    vert_offset = 0.6

    bot_left = [horiz_offset_bottom*width, height]
    bot_right = [(1-horiz_offset_bottom)*width, height]
    top_left = [horiz_offset_top*width, vert_offset*height]
    top_right = [(1-horiz_offset_top)*width, vert_offset*height]

    vertices = np.array([[bot_left, bot_right, top_right, top_left]], dtype=np.int32)
    edges = region_of_interest(edges, vertices)
    plt.figure()
    plt.imshow(edges, cmap='gray')
    mpimg.imsave(report_dir + image_name.split('.')[0] + '_5roi.jpg', edges, cmap = cm.gray)

    # Just plot ROI     
    test_image = np.ones(v.shape)*255
    test_image = region_of_interest(test_image, vertices)
    plt.figure()
    plt.imshow(test_image, cmap='gray')
    mpimg.imsave(report_dir + image_name.split('.')[0] + '_5justroi.jpg', test_image, cmap = cm.gray)

    # Hough line transform
    rho = 1
    theta = np.pi/180
    threshold = 50
    min_line_len = 10
    max_line_gap = 50

    lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    
    line_img = np.zeros((v.shape[0], v.shape[1], 3), dtype=np.uint8)
    # Plot just the raw Hough lines
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(line_img, (x1, y1), (x2, y2), [255, 0, 0], 4)
            
    plt.figure()
    plt.imshow(line_img)
    mpimg.imsave(report_dir + image_name.split('.')[0] + '_6hough.jpg', line_img)

    line_img = np.zeros((v.shape[0], v.shape[1], 3), dtype=np.uint8)
    # Note, draw lines function has been updated to classify lines.
    draw_lines(line_img, np.array(lines), [255, 0, 0] , 10)

    plt.figure()
    plt.imshow(line_img, cmap='gray')
    mpimg.imsave(report_dir + image_name.split('.')[0] + '_7weightedavg.jpg', line_img)

    out_image = weighted_img(line_img, image)
    plt.figure()
    plt.imshow(out_image)

    # Write output to output_dir
    mpimg.imsave(out_dir + image_name.split('.')[0] + '_out.jpg', out_image)
````

````python
## Input: colour image (3 channel)
## Output: colour image with lane lines drawn on in red (3 channel)

def process_image(image):
    # Convert image to greyscale
    image_conv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(image_conv)

    # Threshold greyscale image
    v = (v > 200) * v

    height, width = v.shape

    # Gaussian blur
    v = gaussian_blur(v, 7)

    # Perform dilation to make sure edges are continuous
    v = cv2.dilate(v, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7)), iterations = 1)

    # Canny edge detector
    edges = canny(v, 100, 200)

    # ROI (trapezoid)
    horiz_offset_bottom = 0.05
    horiz_offset_top = 0.35
    vert_offset = 0.6

    bot_left = [horiz_offset_bottom*width, height]
    bot_right = [(1-horiz_offset_bottom)*width, height]
    top_left = [horiz_offset_top*width, vert_offset*height]
    top_right = [(1-horiz_offset_top)*width, vert_offset*height]

    vertices = np.array([[bot_left, bot_right, top_right, top_left]], dtype=np.int32)
    edges = region_of_interest(edges, vertices)

    # Hough line transform
    rho = 1
    theta = np.pi/180
    threshold = 30
    min_line_len = 20
    max_line_gap = 50

    lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
        
    line_img = np.zeros((v.shape[0], v.shape[1], 3), dtype=np.uint8)
    
    # Note, draw lines function has been updated to classify lines.
    draw_lines(line_img, np.array(lines), [255, 0, 0], 10)

    result = weighted_img(line_img, image)

    return result
````
